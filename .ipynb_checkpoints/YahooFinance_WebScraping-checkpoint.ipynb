Below imported are the modules that are being used further in the coding.

```python
from bs4 import BeautifulSoup
from pprint import pprint
import json
import urllib
import requests
import re
import pandas as pd
```

Below code requests the data for harvesting from Yahoo finance website.

```python
top50_html = requests.get("https://finance.yahoo.com/sector/ms_technology").text

top50_soup = BeautifulSoup(top50_html, 'html.parser')

pprint(top50_soup)
```

Below code extracts the stocks mnemonics/symbol corresponding to companys listed in the market.

```python
a_loop_body = top50_soup.find_all('a')

a_tag_data = []

for i in a_loop_body:
    a_tag_data.append(i.text)

pprint(a_tag_data)
```

Below code extracts the data pertaining to TOP 50 companies from the Technological Industry.

```python
company_codes = []
for i in range(len(a_tag_data)):
    if a_tag_data[i] == "Heatmap View":
        for j in range(1,51):
            company_codes.append(a_tag_data[i+j])
        break
print(company_codes)
```

Below code builds the links that are required for harvesting individual attributes of companies. This link differs from company to company.

```python
html_list = []
part_1 = "https://finance.yahoo.com/quote/"
part_2 = "/profile?p="
for code in company_codes:
    html = part_1 + code + part_2 + code
    html_list.append(html)
pprint(html_list)
```

Below code does the following functionalities.

    Retrieves attributes like:
    1) 'Sector' and 'Industry' to which the company belongs to.
    2) Number of 'FUll Time Employees' count
    3) Names of the high-level executives running the company along with their 'Job Title', 'Pay', 'Year Born'.
    
    Builds the json file which stores the multi-layered dictionary level data.
    
    Builds the csv file with the retrieved data.

```python
profile_data = {}
com = 0

for company in company_codes:
    profile_data[company] = {}
    
for link in html_list:

    html_text = requests.get(link).text
    soup = BeautifulSoup(html_text, 'html.parser')
    loop_body = soup.find_all('span')

    tag_data = []
    value = False

    for i in loop_body:
        if i.find_all('span'):
            value = True
            #print(i.text)
        else:
            tag_data.append(i.text)

    name = []
    title = []
    pay = []
    exercised = []
    year_born = []

    for i in range(len(tag_data)):

        if tag_data[i] == "Sector":
            profile_data[company_codes[com]][tag_data[i]] = tag_data[i + 1]
        elif tag_data[i] == "Industry":
            profile_data[company_codes[com]][tag_data[i]] = tag_data[i + 1]
        elif tag_data[i] == "Full Time Employees":
            profile_data[company_codes[com]][tag_data[i]] = tag_data[i + 1]

        if tag_data[i] == "Name":
            for j in range(5,26,5):
                name.append(tag_data[i + j])
            profile_data[company_codes[com]][tag_data[i]] = name
        elif tag_data[i] == "Title":
            for j in range(5,26,5):
                title.append(tag_data[i + j])
            profile_data[company_codes[com]][tag_data[i]] = title
        elif tag_data[i] == "Pay":
            for j in range(5,26,5):
                pay.append(tag_data[i + j])
            profile_data[company_codes[com]][tag_data[i]] = pay
        elif tag_data[i] == "Exercised":
            for j in range(5,26,5):
                exercised.append(tag_data[i + j])
            profile_data[company_codes[com]][tag_data[i]] = exercised
        elif tag_data[i] == "Year Born":
            for j in range(5,26,5):
                year_born.append(tag_data[i + j])
            profile_data[company_codes[com]][tag_data[i]] = year_born
    com += 1

    df = pd.DataFrame.from_dict(profile_data, orient="index")
    df.to_csv("Stock_Market_Attributes.csv")

with open("Stock Market Attributes.json", "w") as json_file:
    json.dump(profile_data, json_file)

pprint(profile_data)
    
```

**Web Scrapping of 50 stock market companies** 
The code below will extract data of the top 50 companies in the Stock Exchange Market from the Yahoo Finance website. 
The data includes the history of every company for the past 5 years from now. Its taking data for every single day.
Based on our requirement, we can change the timestamp accordingly by modifying the variable "days_back" which will take the data ranging from a single day up until our requirement.


```python
import re
from io import StringIO
from datetime import datetime, timedelta

import requests
import pandas as pd


class YahooFinanceHistory:
    timeout = 10
    crumb_link = 'https://finance.yahoo.com/quote/{0}/history?p={0}'
    crumble_regex = r'CrumbStore":{"crumb":"(.*?)"}'
    quote_link = 'https://query1.finance.yahoo.com/v7/finance/download/{quote}?period1={dfrom}&period2={dto}&interval=1d&events=history&crumb={crumb}'

    def __init__(self, symbol, days_back=7):
        self.symbol = symbol
        self.session = requests.Session()
        self.dt = timedelta(days=days_back)

    def get_crumb(self):
        response = self.session.get(self.crumb_link.format(self.symbol), timeout=self.timeout)
        response.raise_for_status()
        match = re.search(self.crumble_regex, response.text)
        if not match:
            raise ValueError('Could not get crumb from Yahoo Finance')
        else:
            self.crumb = match.group(1)

    def get_quote(self):
        if not hasattr(self, 'crumb') or len(self.session.cookies) == 0:
            self.get_crumb()
        now = datetime.utcnow()
        dateto = int(now.timestamp())
        datefrom = int((now - self.dt).timestamp())
        url = self.quote_link.format(quote=self.symbol, dfrom=datefrom, dto=dateto, crumb=self.crumb)
        response = self.session.get(url)
        response.raise_for_status()
        return pd.read_csv(StringIO(response.text), parse_dates=['Date'])
```

Below code describes how we are scrapping the URL of "yahoo finance" to get the top 50 companies.

```python
from bs4 import BeautifulSoup
from pprint import pprint
import urllib
import requests
import re

top50_html = requests.get("https://finance.yahoo.com/sector/ms_technology").text
#print(html_text)

top50_soup = BeautifulSoup(top50_html, 'html.parser')
print(type(top50_soup))
pprint(top50_soup)
```

The next step was to scrap out the different stock exchange market along with the company codes.

```python
a_tag_list = ['a']
a_loop_body = top50_soup.find_all(a_tag_list)

#pprint(a_loop_body)
a_tag_data = []

for i in a_loop_body:
    a_tag_data.append(i.text)

pprint(a_tag_data)
```

The below code takes out the company codes from the above output for further processing.

```python
company_codes = []
for i in range(len(a_tag_data)):
    if a_tag_data[i] == "Heatmap View":
        for j in range(1,51):
            company_codes.append(a_tag_data[i+j])
        break
print(company_codes)
```

The below code we are scrapping out the data from each companies statistics page and building a csv file which contains the top 50 companies codes along with their different attributes like: 1.Enterprise Value 2.Quaterly Growth 3.Forward P/E 4.PEG Ratio (5 yr expected) 5.Return on Assets 6.Quarterly Revenue Growth 7.EBITDA 8.Diluted EPS 9.Total Debt/Equity 10.Current Ratio, which gives us a clear description about the top 50 companies in the stock exchange market.

```python
import requests, re, json , pprint


pr = re.compile(r'root\.App\.main = (.*);')
tickers = ['AAPL', 'MSFT', 'TSM','INTC', 'CSCO','ORCL', 'SAP','ADBE','NVDA','ACN', 'AVGO','IBM', 'ASML', 'TXN','QCOM', 'FIS', 'SNE', 'INTU','VMW','MU', 'AMAT', 'NOW','AMD', 'INFY', 'ADI','ADSK', 'LRCX','DELL','CTSH','NXPI', 'APH', 'TEL','HPQ','CAJ','MSI', 'FLT','KLAC', 'FTV-PA', 'FTV','MCHP', 'XLNX','STM','CRM','UBER','SHOP','WDAY','ERIC','TEAM','SPLX']
result = {}

with requests.Session() as s:

    for ticker in tickers:
        rp = s.get('https://finance.yahoo.com/quote/{}/key-statistics?p={}'.format(ticker,ticker))
        data = json.loads(pr.findall(rp.text)[0])
        key_stats = data['context']['dispatcher']['stores']['QuoteSummaryStore']
       #pprint(data['context']['dispatcher']['stores']['QuoteSummaryStore'])
        if (key_stats['defaultKeyStatistics'] != {}):
            if(key_stats['summaryDetail'] != {}):
                if(key_stats['financialData'] != {}):
                    ress = {
                        'Enterprise Value' : key_stats['defaultKeyStatistics']['enterpriseValue']
                        ,'Quaterly Growth' : key_stats['defaultKeyStatistics']['earningsQuarterlyGrowth']
                        ,'Forward P/E' : key_stats['summaryDetail']['forwardPE']
                        ,'PEG Ratio (5 yr expected)' : key_stats['defaultKeyStatistics']['pegRatio']
                        , 'Return on Assets' : key_stats['financialData']['returnOnAssets']
                        , 'Quarterly Revenue Growth' : key_stats['financialData']['revenueGrowth']
                        , 'EBITDA' : key_stats['financialData']['ebitda']
                        , 'Diluted EPS' : key_stats['defaultKeyStatistics']['trailingEps']
                        , 'Total Debt/Equity' : key_stats['financialData']['debtToEquity']
                        , 'Current Ratio' :  key_stats['financialData']['currentRatio']
                    }
                    result[ticker] = ress
                else:
                    ress = {
                        'Enterprise Value' : 'N/A'
                        ,'Trailing P/E' : 'N/A'
                        ,'Forward P/E' : 'N/A'
                        ,'PEG Ratio (5 yr expected)' : 'N/A'
                        , 'Return on Assets' : 'N/A'
                        , 'Quarterly Revenue Growth' : 'N/A'
                        , 'EBITDA' : 'N/A'
                        , 'Diluted EPS' : 'N/A'
                        , 'Total Debt/Equity' : 'N/A'
                        , 'Current Ratio' :  'N/A'
                    }
                    result[ticker] = ress
pprint.pprint(result)

import csv
import pandas as p
df = p.DataFrame.from_dict(result, orient="index")
df.to_csv('Statistics.csv')
```
